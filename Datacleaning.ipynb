```python
import pandas as pd
df = pd.read_csv('shows.csv')
print("Data:")
print(df.head())

# Standardize column names 
df.columns = [col.lower().replace(' ', '_') for col in df.columns]

df = df.dropna()

df = df.drop_duplicates()

df['title'] = df['title'].str.strip()

print("Cleaned Data:")
print(df.head())

df.to_csv('cleaned_movies_dataset.csv', index=True)
```
